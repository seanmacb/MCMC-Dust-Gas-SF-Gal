```python
import pandas as pd
import operator
import numpy as np
import scipy.stats as sps

import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import cm

import matplotlib as mpl
mpl.use('TkAgg')

import emcee
import pymc3 as pm

from astropy.table import Table
import astropy.io as asio
from astropy.io import fits
import corner

import io
from google.colab import files

import matplotlib.gridspec as gridspec

%matplotlib inline
```

```python
def nanmask(array):
    mask = (np.isnan(array) == False)
    newarray = array[mask]
    return newarray, mask

def seanscat(x,y):
    xlab = input("What would you like the x label to be?")
    ylab = input("What would you like the y label to be?")
    pointsize = input("What would you like the size of the scatter points to be?")
    plt.scatter(x,y,s=int(pointsize))
    plt.xlim(0.9*min(x),1.1*max(x))
    plt.ylim(0.9*min(y),1.1*max(y))
    plt.grid()
    plt.xlabel(xlab)
    plt.ylabel(ylab)
    plt.title(xlab+" vs. "+ylab)
    plt.show()
    
def reducer(df,keycol,key=np.inf):
    delindex=[]
    for i in range(len(keycol)):
        if keycol[i]!=key:
            delindex.append(i)
    return df.drop(delindex)

def reducer2(df,keycol,key=np.inf):
    delindex=[]
    for i in range(len(keycol)): # if key = keycol value
        if keycol[i]==key:
            delindex.append(i)
    return df.drop(delindex)

def reducer3(df,keycol):
    delindex=[]
    for i in range(len(keycol)):
        if np.isnan(keycol[i])==True:
            delindex.append(i)
    return df.drop(delindex)

def residMCMC(sig_y_obs,m,sig_m,x,sig_x,sig_b):
    sig_mx = np.abs(m*x)*np.sqrt((sig_m/m)**2+(sig_x/x)**2)
    sig_mxpb = np.sqrt(sig_mx**2+sig_b**2)
    return np.sqrt((sig_y_obs)**2+(sig_mxpb)**2)

from matplotlib.pyplot import figure
BDecLabel = "Log(H\u03B1/H\u03B2)"
LCO10Label = "Log(L CO $_{1-0}$)"
LCO21Label = "Log(L CO $_{2-1}$)"
MDustLabel = "Log(M$_{Dust}$/M$_*$)"
MH2Label = "Log(M H$_2$/M$_*$)"
StarMassLabel = "Log(M$_{Star}$/M$_*$)"
SFRLabel = "SFR"
GasFracLabel = "GasFrac"
N2HALabel = "Log(N$_2$/H\u03B1"
```

```python
MPAMaster = pd.read_csv("MPAMaster.csv") # mpa master set
xColdDF = pd.read_csv("xColdFilt.csv") # xcold data
dfJingle = pd.read_csv("JINGLEMASTERDEC19.csv") # Jingle data
# print(dfJingle.columns.values)

m_MDust = 2.67 # dust - reran!
b_MDust = 6.11
m_sd_MDust = .3
b_sd_MDust = .2

m_LCO10 = 6.25 # LCO 1-0 - reran!
b_LCO10 = 4.34
m_sd_LCO10 = .5
b_sd_LCO10 = .3

m_MH2 = 4.27 # MH2 - reran!
b_MH2 = 6.29
m_sd_MH2 = .42
b_sd_MH2 = .26

m_LCO21 = 4.34 # LCO 2-1
b_LCO21 = 5.53
m_sd_LCO21 = .41
b_sd_LCO21 = .26

xColdDF["LOGLCO21"] = np.log10(xColdDF["LCO21"])
xColdDF["LOGLCO21_ERR"] = xColdDF["LCO21_ERR"]/(xColdDF["LCO21"]*np.log(10))

xColdDF=reducer2(xColdDF,np.asarray(xColdDF["LOGLCO21"]),0)
xColdDF.to_csv("xColdnoLCO21.csv") # uses only galaxies with LCO 2-1 transition detection

xColdDF["ResidMH2"] = xColdDF["LOGMH2"]-(xColdDF["BALMER"]*m_MH2+b_MH2) # MH2
xColdDF["ResidMH2_err"] = residMCMC(xColdDF["LOGMH2_ERR"],m_MH2,m_sd_MH2,xColdDF["BALMER"],xColdDF["BALMER_ERR"],b_sd_MH2)

xColdDF["ResidLCO21"] = xColdDF["LOGLCO21"]-(xColdDF["BALMER"]*m_LCO21+b_LCO21) # LCO 2-1

xColdDF["ResidLCO10"] = xColdDF["LOGLCO"]-(xColdDF["BALMER"]*m_LCO10+b_LCO10) # LCO 1-0
xColdDF["ResidLCO10_err"] = residMCMC(xColdDF["LOGLCO_ERR"],m_LCO10,m_sd_LCO10,xColdDF["BALMER"],xColdDF["BALMER_ERR"],b_sd_LCO10)

xColdDF["ResidGasFrac"] = (xColdDF["LOGMH2MS"])-((xColdDF["BALMER"]*m_MH2+b_MH2)/xColdDF["MASS"]) # gas frac

dfJingle["ResidMDust"] = (dfJingle["logMDust"])-(dfJingle["MPA LBdec"]*m_MDust+b_MDust) # dust mass
dfJingle["ResidMDust_err"] = residMCMC(np.sqrt(dfJingle["logMDust uperr"]**2+dfJingle["logMDust lowerr"]**2),m_MDust,m_sd_MDust,dfJingle["MPA LBdec"],dfJingle["MPA LBdec err"],b_sd_MDust)

delindex = []
for i in range(len(dfJingle["N2HA"])):
    if dfJingle["AGNCLASS"][i]!=1:
        delindex.append(i)
dfJingle = dfJingle.drop(delindex) # removing non SFing galaxies

# print(MPAMaster.columns.values)
print(xColdDF.columns.values)
# print(dfJingle.columns.values)
```

```python
# Plot figure with subplots of different sizes
fig = plt.figure(1)
# set up subplot grid
gridspec.GridSpec(4,1)

# large subplot
plt.subplot2grid((4,1), (0,0), rowspan=2)
plt.locator_params(axis=BDecLabel, nbins=5)
plt.locator_params(axis=MDustLabel, nbins=5)
# plt.title('Comparison )
plt.title(MDustLabel)
plt.xlabel(BDecLabel)
plt.ylabel(MDustLabel)
plt.scatter(dfJingle["MPA LBdec"],dfJingle["logMDust"],c="blue",s=2,label="JINGLE")
plt.errorbar(dfJingle["MPA LBdec"],dfJingle["logMDust"],xerr=dfJingle["MPA LBdec err"],yerr=[dfJingle["logMDust lowerr"],dfJingle["logMDust uperr"]],c="blue",elinewidth=.5,linestyle="none")
plt.plot([0,1.5],[b_MDust,b_MDust+m_MDust*1.5],c="black",linestyle="--",lw=0.5,label="MCMC Linear Fit")
plt.legend()
plt.xlim(0.3,1.1)
plt.ylim(7,9)
plt.grid()

plt.subplot2grid((4,1), (2,0))
plt.locator_params(axis="Inclination", nbins=5)
plt.locator_params(axis=MDustLabel, nbins=5)
# plt.title('Comparison )
plt.xlabel("Inclination")
plt.ylabel("Residual "+MDustLabel)
plt.scatter(dfJingle["INCL"],dfJingle["ResidMDust"],c="blue",s=2)
plt.errorbar(dfJingle["INCL"],dfJingle["ResidMDust"],yerr=dfJingle["ResidMDust_err"],c="blue",linestyle="none",elinewidth=0.5)
plt.plot([0,90],[0,0],c="black",linestyle="--")
plt.xlim(0,90)
plt.ylim(-1,1)
plt.grid()

plt.subplot2grid((4,1), (3,0))
plt.locator_params(axis="Metallicity", nbins=5)
plt.locator_params(axis=MDustLabel, nbins=5)
# plt.title('Comparison )
plt.xlabel("Log(N$_2$/H\u03B1)")
plt.ylabel("Residual "+MDustLabel)
plt.scatter(dfJingle["N2HA"],dfJingle["ResidMDust"],c="blue",s=2)
plt.errorbar(dfJingle["N2HA"],dfJingle["ResidMDust"],yerr=dfJingle["ResidMDust_err"],c="blue",linestyle="none",elinewidth=0.5)
plt.plot([-1,1],[0,0],c="black",linestyle="--")
plt.xlim(-0.8,0.1)
plt.ylim(-1,1)
plt.grid()

# fit subplots and save fig
# fig.tight_layout()
fig.set_size_inches(w=6,h=15)
fig_name = 'MDust_INCLRES.png'
fig.savefig(fig_name,dpi=300)
```
