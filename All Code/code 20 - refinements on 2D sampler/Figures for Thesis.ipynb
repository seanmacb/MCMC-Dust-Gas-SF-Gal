```python
import matplotlib.pyplot as plt
import numpy as np

h = 6.626e-34
c = 3.0e+8
k = 1.38e-23

def planck(wav, T):
    a = 2.0*h*c**2
    b = h*c/(wav*k*T)
    intensity = a/ ( (wav**5) * (np.exp(b) - 1.0) )
    return intensity

# generate x-axis in increments from 1nm to 3 micrometer in 1 nm increments
# starting at 1 nm to avoid wav = 0, which would result in division by zero.
wavelengths = np.arange(1e-9, 3e-6, 1e-9) 

intensity4 = planck(wavelengths, 4000.)
intensity5 = planck(wavelengths, 5000.)
intensity6 = planck(wavelengths, 6000.)
intensity7 = planck(wavelengths, 7000.)

nu = c/wavelengths

# plt.hold(True)
plt.plot(wavelengths*10**9, intensity4, 'r-',label="4000 K") 
plt.plot(wavelengths*10**9, intensity5, 'g-',label="5000 K")
plt.plot(wavelengths*10**9, intensity6, 'b-',label="6000 K")
plt.plot(wavelengths*10**9, intensity7, 'k-',label="7000 K")
plt.ylabel("Luminosity [$W/m^2$]")
plt.xlim(0,3e+3)
plt.xlabel("$\lambda$ [$nm$]")
# plt.xlabel("Hz")
plt.ylim(0,7e+13)
plt.legend()
plt.savefig("PlancksLaw.png",dpi=300)
plt.show()
```
