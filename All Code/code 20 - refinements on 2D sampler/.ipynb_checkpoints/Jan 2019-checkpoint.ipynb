```python
def nanmask(array):
    mask = (np.isnan(array) == False)
    newarray = array[mask]
    return newarray, mask

def seanscat(x,y):
    xlab = input("What would you like the x label to be?")
    ylab = input("What would you like the y label to be?")
    pointsize = input("What would you like the size of the scatter points to be?")
    plt.scatter(x,y,s=int(pointsize))
    plt.xlim(0.9*min(x),1.1*max(x))
    plt.ylim(0.9*min(y),1.1*max(y))
    plt.grid()
    plt.xlabel(xlab)
    plt.ylabel(ylab)
    plt.title(xlab+" vs. "+ylab)
    plt.show()
    
def reducer(df,keycol,key=np.inf):
    delindex=[]
    for i in range(len(keycol)):
        if keycol[i]!=key:
            delindex.append(i)
    return df.drop(delindex)

def reducer2(df,keycol,key=np.inf):
    delindex=[]
    for i in range(len(keycol)): # if key = keycol value
        if keycol[i]==key:
            delindex.append(i)
    return df.drop(delindex)

def reducer3(df,keycol):
    delindex=[]
    for i in range(len(keycol)):
        if np.isnan(keycol[i])==True:
            delindex.append(i)
    return df.drop(delindex)

def residMCMC(sig_y_obs,m,sig_m,x,sig_x,sig_b):
    sig_mx = np.abs(m*x)*np.sqrt((sig_m/m)**2+(sig_x/x)**2)
    sig_mxpb = np.sqrt(sig_mx**2+sig_b**2)
    return np.sqrt((sig_y_obs)**2+(sig_mxpb)**2)

def linfit(m,x,b):
    return m*x+b

def percSplit(splitArr,lowlim=15.87,uplim=84.13): # defaults to 1sigma levels
    upperval = np.percentile(splitArr,uplim)
    lowerval = np.percentile(splitArr,lowlim)
    upperlist,lowerlist = [],[]
    for i in range(len(splitArr)):
        if np.asarray(splitArr)[i]>upperval:
            upperlist.append(i)
        elif np.asarray(splitArr)[i]<lowerval:
            lowerlist.append(i)
    return upperval, lowerval, upperlist,lowerlist

from matplotlib.pyplot import figure
BDecLabel = "$\log$(H\u03B1/H\u03B2)"
LCO10Label = "$\log(L_{CO_{1-0}})$ [K km s$^{-1}$ pc$^2$]"
LCO21Label = "$\log$($L_{CO_{2-1}}$)"
MDustLabel = "$\log$(M$_{Dust})$[M$_\odot$]"
MH2Label = "$\log$($M_{H_2})$[M$_\odot$]"
StarMassLabel = "$\log$(M$_{Star}$) [M$_\odot$]"
SFRLabel = "$\log$(SFR) [$M_\odot$ yr$^{-1}$]"
GasFracLabel = "$\log (f_{H_2})$"
N2HALabel = "$\log$(N$_2$/H\alpha)"
```

```python
MPAMaster = pd.read_csv("MPAMaster.csv") # mpa master set
# MPAMaster = MPAMaster.dropna()
xColdDF = pd.read_csv("xColdFilt.csv") # xcold data
dfJingle = pd.read_csv("JINGLEMASTERDEC19.csv") # Jingle data
dfJingle = dfJingle.dropna()

m_MDust = 2.59 # dust
b_MDust = 6.15
m_sd_MDust = .3
b_sd_MDust = .2
f_MDust = 0.28

m_LCO10 = 5.89 # LCO 1-0
b_LCO10 = 4.59
m_sd_LCO10 = .53
b_sd_LCO10 = .33
f_LCO10 = 0.44

m_MH2 = 4.27 # MH2
b_MH2 = 6.29
m_sd_MH2 = .42
b_sd_MH2 = .26
f_MH2 = 0.31

# m_LCO21 = 4.34 # LCO 2-1
# b_LCO21 = 5.53
# m_sd_LCO21 = .41
# b_sd_LCO21 = .26

# xColdDF["LOGLCO21"] = np.log10(xColdDF["LCO21"])
# xColdDF["LOGLCO21_ERR"] = xColdDF["LCO21_ERR"]/(xColdDF["LCO21"]*np.log(10))

# xColdDF=reducer2(xColdDF,np.asarray(xColdDF["LOGLCO21"]),0)
# xColdDF.to_csv("xColdnoLCO21.csv") # uses only galaxies with LCO 2-1 transition detection

# MPAMaster["ResidStarMass"] = MPAMaster["StarMass"]-(MPAMaster["BDec"]*m_S+b_S)

# delindex = []
# for i in range(len(dfJingle["N2HA"])):
#     if dfJingle["AGNCLASS"][i]!=1:
#         delindex.append(i)
# dfJingle = dfJingle.drop(delindex) # removing non SFing galaxies

# # delindex=[]
# # for i in range(len(dfJingle)):
# #     if np.asarray(dfJingle["N2HA"])[i]<-20:
# #         delindex.append(i)
# # dfJingle=dfJingle.drop(delindex) # removing bad values in jingle data

# delindex=[]
# for i in range(len(xColdDF)):
#     if np.asarray(xColdDF["LOGMH2"])[i]<4:
#         delindex.append(i)
# xColdDF=xColdDF.drop(delindex)

# tmparr=[]
# for i in range(len(MPAMaster)):
#     if MPAMaster["12+LOG OH"][i]<7:
#         tmparr.append(i)
# MPAMaster = MPAMaster.drop(tmparr)

xColdDF["ResidMH2"] = xColdDF["LOGMH2"]-(xColdDF["BALMER"]*m_MH2+b_MH2) # MH2
xColdDF["ResidMH2_err"] = residMCMC(xColdDF["LOGMH2_ERR"],m_MH2,m_sd_MH2,xColdDF["BALMER"],xColdDF["BALMER_ERR"],b_sd_MH2)

# xColdDF["ResidLCO21"] = xColdDF["LOGLCO21"]-(xColdDF["BALMER"]*m_LCO21+b_LCO21) # LCO 2-1

xColdDF["ResidLCO10"] = xColdDF["LOGLCO"]-(xColdDF["BALMER"]*m_LCO10+b_LCO10) # LCO 1-0
xColdDF["ResidLCO10_err"] = residMCMC(xColdDF["LOGLCO_ERR"],m_LCO10,m_sd_LCO10,xColdDF["BALMER"],xColdDF["BALMER_ERR"],b_sd_LCO10)

xColdDF["ResidGasFrac"] = (xColdDF["LOGMH2MS"])-((xColdDF["BALMER"]*m_MH2+b_MH2)/xColdDF["MASS"]) # gas frac

dfJingle["ResidMDust"] = (dfJingle["logMDust"])-(dfJingle["MPA LBdec"]*m_MDust+b_MDust) # dust mass
dfJingle["ResidMDust_err"] = residMCMC(np.sqrt(dfJingle["logMDust uperr"]**2+dfJingle["logMDust lowerr"]**2),m_MDust,m_sd_MDust,dfJingle["MPA LBdec"],dfJingle["MPA LBdec err"],b_sd_MDust)

MDust_rho,MDust_pval = sps.spearmanr(dfJingle["MPA LBdec"],dfJingle["logMDust"])
MH2_rho,MH2_pval = sps.spearmanr(xColdDF["BALMER"],xColdDF["LOGMH2"])
LCO_rho,LCO_pval = sps.spearmanr(xColdDF["BALMER"],xColdDF["LOGLCO"])

MPAMaster["MH2_MCMC"] = m_MH2*MPAMaster["BDec"]+b_MH2
MPAMaster["LCO10_MCMC"] = m_LCO10*MPAMaster["BDec"]+b_LCO10

MPAMaster["MDust_MCMC"] = m_MDust*MPAMaster["BDec"]+b_MDust

MPAMaster["GasFrac_MCMC"] = np.log10((10**MPAMaster["MH2_MCMC"])/(10**MPAMaster["StarMass"]))
```

```python
x= np.linspace(-1,1,num=200)

def gaussian(x,mu,sig):
    return (1/(sig*np.sqrt(2*np.pi)))*np.exp((-1/2)*((x-mu)/sig)**2)

y1 = gaussian(x,0.1,0.1) # likelihood
y2 = gaussian(x,-0.15,0.5) # prior

plt.plot(x,y1,label="Likelihood",lw=0.8)
plt.plot(x,y2,label="Prior",lw=0.8)
plt.plot(x,y1*y2,label="Posterior",lw=0.8)
plt.legend()
plt.grid()

plt.xlim(-1,1)
plt.yticks([])

plt.savefig("PriorsAndPosteriors.png",dpi=300)

plt.show()
```
