```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.gridspec as gridspec
```

```python
def eq1(y,R_v):
    return aEQ(y)+bEQ(y)/R_v

def aEQ(y):
    return (1+0.17699*y-0.50447*y**2-0.02427*y**3+0.72085*y**4+0.01979*y**5-0.7753*y**6+0.32999*y**7)

def bEQ(y):
    return (1.41338*y+2.28305*y**2+1.07233*y**3-5.38434*y**4-0.62251*y**5+5.30260*y**6-2.09002*y**7)
```

Here we're trying to show the relationship between $R_v$, $E(B-V)$, and the modified balmer decrement in galaxies. $BD_0=\frac{H\alpha}{H\beta}=2.86$ from Osterbrock 1989, although that refers to gaseous nebulae and not galaxies.

$A_v=R_v \times E(B-V)$

$BD(R_v,E(B-V)) = BD_0 \times 10^{0.4A_v\times(A_\beta-A_\alpha)} \equiv BD(R_v,E(B-V)) = BD_0 \times 10^{0.4R_vE(B-V)\times(A_\beta-A_\alpha)}$

$\lambda_\alpha (656.56nm) > \lambda_\beta (486.13nm) \rightarrow E\propto \frac{1}{\lambda}$, therefore $E_\alpha < E_\beta$ & $A_\alpha < A_\beta$. This means that the balmer decrement will be higher in environments with higher extinction. This is evidence by the plot above

```python
BD0 = 2.86

Rv_range = np.linspace(1,6,num=500) 
EB_V_range = np.linspace(0,0.3,num=500) #Ranges for fixed plots

HAlpha = 0.65628 # Wavelength in microns
HBeta = 0.48613

yHAlpha = (1/HAlpha)-1.82 # eqn 3a, an adjustment made by CCM 1989
yHBeta = (1/HBeta)-1.82

# Extinction calculations

A_alpha = eq1(yHAlpha,Rv_range) # deriving how the extinction at H-alpha and H-beta at varying Rv according to CCM 1989
A_beta = eq1(yHBeta,Rv_range) # these quantities are based on the ranges defined above, not to be confuse with the gridspaces defined below!

# converting to a meshgrid

Rv_mesh, EB_mesh = np.meshgrid(Rv_range,EB_V_range)

A_alpha_mesh = eq1(yHAlpha,Rv_mesh)
A_beta_mesh = eq1(yHBeta,Rv_mesh) # recall equation 1 from CCM

def ModifiedBD(Rv,EB_V,A_alpha,A_beta):
    return BD0*10**(0.4*Rv*EB_V*(A_beta-A_alpha)) # the main equation we derived -> note the multiplication by A_v within the exponent has been replaced with the appropriate extinction relation
                    
GridBD = ModifiedBD(Rv_mesh,EB_mesh,A_alpha_mesh,A_beta_mesh) # getting gridspace of the 
```

```python
fig = plt.figure(figsize=[5.5,4.76])

plt.scatter(Rv_mesh,EB_mesh,c=np.log10(GridBD))

plt.xlabel("$R_v$")
plt.ylabel("$E(B-V)$")
plt.colorbar()
plt.text(7.3,0.12, "log(Ha/Hb)", rotation=-90)

plt.xlim(1,6)
plt.ylim(0,0.3)
# plt.clim(0.5,1)

plt.savefig("Extinction Colormap.png",dpi=300)

plt.show()
```

```python
for i in range(1,7):
    plt.scatter(Rv_range,np.log10(ModifiedBD(Rv_range,i,A_alpha,A_beta)),label="$E(B-V)=$"+str(i),s=5)

plt.ylabel("log(Ha/Hb)")
plt.xlabel("$R_v$")
plt.xlim(1,6)
plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc='lower left',
           ncol=3, mode="expand", borderaxespad=0.)
plt.grid()
plt.savefig("RVvBDec.pdf",dpi=300)

plt.show()
```

```python
for i in range(1,7):
    plt.scatter((EB_V_range),np.log10(ModifiedBD(i,EB_V_range,A_alpha,A_beta)),label="$R_v=$"+str(i),s=5)

plt.ylabel("log(Ha/Hb)")
plt.xlabel("$E(B-V)$")
plt.xlim(0,5)
plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc='lower left',
           ncol=3, mode="expand", borderaxespad=0.)
plt.grid()
plt.savefig("EB_VvBDec.pdf",dpi=300)

plt.show()
```

Final formula: $M(HI+H_2) [kg] = 5.8\times10^{21}\times m_H\times E(B-V)\times F_{correct}\times D^2\times a$ with $F_{correct}=\left(\frac{\pi}{180\times 3600}\right)^2\times (3.086\times 10^{24})^2 [cm^2]$, a [arcsec$^2$], D [Mpc], $m_H$ [kg], and E(B-V) [mag]

```python
Bohlin_n=5.8*10**21 # original bohlin factor
m_H = 1.6735575*10**(-27) # mass of hydrogen in kg

# calculating correction factor for 1Mpc and a=1 arcsec^2
A_correct = (np.pi/(180*3600))**2*(3.086*10**24)**2


def MassH(EBV,D,A): # Full equation
    return Bohlin_n*A_correct*m_H*EBV*(D**2)*A

# making ranges
EB_V_range = np.linspace(0.009,0.27,num=6) # 0 to 5 E(B-V)
D_range = np.linspace(0.778,300,num=500) # 0 to 300 Mpc
A_range=np.linspace(0,20000,num=500) # 0 to 2*10^4 arcsec^2

# Making meshgrid
D_mesh, A_mesh = np.meshgrid(D_range,A_range)

solarMass = 1.989*10**30
```

```python
# Plot figure with subplots of different sizes
fig = plt.figure(1,figsize=(24,16))
# set up subplot grid
gridspec.GridSpec(2,3)

for i in range(len(EB_V_range)):
    plt.subplot2grid((2,3), (i//3,i%3))
    
    MH_2 = MassH(EB_V_range[i],D_mesh,A_mesh)/solarMass # calculating MH2 for each value of EB-V
    logMH_2 = np.log(MH_2)
    plt.scatter(A_mesh,D_mesh,c=logMH_2)
#     contours = plt.contourf(A_mesh,D_mesh,logMH_2,100, cmap='plasma')
    plt.colorbar()
#     plt.clabel(contours, inline=True, fontsize=8)
#     plt.imshow(logMH_2, origin='lower', cmap='plasma', alpha=0.5)
#     plt.yscale("log")
#     plt.xscale("log")
    
    plt.axis(aspect='image')
    plt.xlabel("Angular area [arcsec$^2$]")
    plt.ylabel("Distance [Mpc]")
    plt.xlim(np.amin(A_mesh),np.amax(A_mesh))
    plt.ylim(np.amin(D_mesh),np.amax(D_mesh))
    plt.text(24000,135,"$\log({M_{H_2}}/{M_\odot})$",rotation=-90)
    plt.grid()
    plt.title("E(B-V)="+'{:04.2f}'.format((EB_V_range[i])))

plt.savefig("H2_EBV.png",dpi=300)
plt.show()
```
